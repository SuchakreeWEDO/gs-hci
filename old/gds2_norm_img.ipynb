```python
import matplotlib.pyplot as plt
import numpy as np

folder_path = r"D:\3d-reconstruction\datasets\media_sony_fixediso\depth"

from os import listdir
from os.path import isfile, join
all_files = [join(folder_path, f) for f in listdir(folder_path) if isfile(join(folder_path, f))]

print(all_files)
```

```python
resolution = (1600, 1064) # get size from rendered depth map
print(resolution)
```

```python
# see comparison between original depth image and norm image

from PIL import Image, ImageOps
from utils.general_utils import PILtoTorch
import torchvision.transforms as T
import torch

for file in all_files:

    if "0100" in file:

        mono_depth = Image.open(file)
        mono_depth = ImageOps.grayscale(mono_depth) 

        mono_depth_res = PILtoTorch(mono_depth, resolution)
        print("mono_depth_res", mono_depth_res.shape, mono_depth_res.min(), mono_depth_res.max())
        transform = T.ToPILImage()
        img = transform(mono_depth_res)
    
        img.show()


        mono_depth_norm = ( mono_depth_res - torch.min(mono_depth_res) ) / ( torch.max(mono_depth_res) - torch.min(mono_depth_res) )
        print("mono_depth_norm", mono_depth_norm.shape, mono_depth_norm.min(), mono_depth_norm.max())
        transform = T.ToPILImage()
        img = transform(mono_depth_norm)
        img.show()
```
